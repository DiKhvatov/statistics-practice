## 0. From the beginning

Напомним постановку задачи. Пусть на нас с неба свалился набор наблюдений $\mathbf X$ --- это может быть выборка, набор независимых выборок и т.д. Пусть известно, что его распределение $\mathsf{P}$ лежит в некотором большом множестве $\mathcal P$, но мы предполагаем, что имеет место несколько более частная ситуация: распределение $\mathbf X$ на самом деле лежит в $\mathcal P_0 \subset \mathcal P$. По данным мы хотим понять, насколько это предположение состоятельно: либо они слишком неправдоподобно выглядят для $\mathcal P_0$, либо оснований для сомнений нет. Такого рода предположения называют *статистической гипотезой* (пишут $H_0\colon \mathsf{P} \in \mathcal{P}_0$). Зачастую ей противопоставляют другую, *альтернативную* гипотезу $H_1$, в то время как $H_0$ называют *основной (нулевой)*, но далее мы будем для простоты считать, что $H_1$ имеет вид $\mathcal{P} \setminus \mathcal{P}_0$.

*Примеры:*
- $(X_1, \ldots, X_n)$ и $(Y_1, \ldots, Y_m)$ --- две независимые выборки. Правда ли, что $\mathsf{E} X_1 = \mathsf{E} Y_1$?
- $(X_1, Y_1), \ldots, (X_n, Y_n)$ --- выборка из случайных векторов. Можно ли считать, что $X_i$ и $Y_i$ независимы?
- $X_1, \ldots, X_n$ --- какая-то выборка. Верно ли, что она пришла из нормального распределения?
- $X_1, \ldots, X_n$ --- выборка из распределения $\mathrm{Bern}(p)$ (например, броски монеты). Можно ли считать её симметричной (т.е. $p=1/2$)?

Для проверки гипотезы строят множество $R$: если выборка в него попала, то гипотеза отвергается, иначе --- нет. Обычно такие множества имеют вид
$$
R_c = \{\mathbf x\colon T(\mathbf x) > c\},
$$
где $T(\mathbf X)$ --- некоторая статистика, называемая *статистикой критерия*. Таким образом, она является прибором для измерения соответствия нулевой гипотезе: если она не очень большая, то повода для беспокойства нет и $H_0$ скорее всего верна; иначе, если её значения зашкаливает и больше некоторого порога, то это является серьёзным основанием для отвержения $H_0$. 

При проверке $H_0$ потенциально мы можем ошибиться: либо отвергнуть $H_0$, когда она на самом деле верна (*ошибка I рода*), либо наоборот, принять её, хотя она неверна (*ошибка II рода*). Минимизировать вероятности обеих ошибок не представляется возможным. Так как обычно ошибка I рода менее желательна, то в первую очередь минимизируют её. Для этого выбирают какую-то маленькую константу $\alpha$ и подбирают порог $c$ так, чтобы 
$$
\forall \mathsf{P}_0 \in \mathcal{P}_0\colon \mathsf{P}_0(\mathbf X \in R_c) \le \alpha.
$$
В таком случае говорят, что критерий обладает уровнем значимости $\alpha$. По этой причине стараются подбирать $T(\mathbf X)$ так, чтобы её точное (или хотя бы асимптотическое) распределение было одним и тем же для любых $\mathsf{P}_0 \in \mathcal{P}_0$: так проще контролировать уровень значимости.

### 0.1 Pvalue

Для удобства проверки гипотез используют понятие *pvalue*. Пусть для каждого $\alpha \in (0; 1)$ у нас имеется критерий $R_{\alpha}$ с ошибкой I рода $\alpha$. Pvalue -- это статистика, равная
$$
pvalue(\mathbf X) = \inf\{\alpha\colon \mathbf X \in R_{\alpha}\},
$$
то есть это наименьший уровень значимости, при котором при имеющихся данных мы бы отвергли нулевую гипоетезу. По определению, если мы умеем считать pvalue, нам достаточно сравнить его с взятым нами уровнем значимости: если pvalue оказалось меньше, то гипотеза отвергается.

Несложно понять (поймите это!), что для критериев вида $R_{\alpha} = \{\mathbf x\colon T(\mathbf x) > c_{\alpha}\}$ pvalue можно посчитать как
$$
pvalue(\mathbf x) = \mathsf{P}_0(T(\mathbf X) > T(\mathbf x)) = 1 - F_{T(\mathbf X)}(T(\mathbf x)),
$$
то есть pvalue --- это вероятность пронаблюдать значение статистики критерия более экстремальное, чем оно есть на самом деле, при верности $H_0$. Таким образом, если мы знаем распределение статистики, то достаточно применить к ней её же функцию распределения, после чего вычесть её из 1. В библиотечных реализациях pvalue считается автоматически.

*Пример:* Пусть $X_1, \ldots, X_n$ --- выборка из $\mathcal{N}(a, 1)$, где $a \in \mathbb{R}$ --- неизвестный параметр. Проверим гипотезу $H_0\colon a = 0$. Для этого логично использовать статистику $T(\mathbf X) = \overline{\mathbf X}$, тем более при верности $H_0$ мы знаем её распределение --- $\mathcal{N}(0, 1/n)$. Чтобы хорошо отлавливать случаи как положительных $a$, так и отрицательных, стоить отклонять гипотезу при больших значениях модуля $T(\mathbf X)$. Таким образом, возьмём критерий $\{\mathbf x\colon |\overline{\mathbf x}| > z_{1-\alpha/2}\}$

```python
x = sps.norm.rvs(random_state=10538)
stat = abs(x.mean())
print(f"Статистика критерия: {stat}")
pvalue = 2 * sps.norm.sf(stat)
print(f"Pvalue: {pvalue}")
```

Если бы мы проверяли гипотезу на уровне значимости 0.05, то результат для нас был бы значимым (pvalue меньше 0.05), и гипотезу пришлось бы отвергнуть. Были бы мы были чуть более консервативными, более боязливыми для совершения ошибки I рода и взяли бы уровень значимости 0.01, то отклонение от нулевой гипотезы показалось бы нам не таким критичным (pvalue больше 0.01), поэтому гипотеза бы принялась. Поэтому важно фиксировать уровень значимости заранее, чтобы не подгоняться под результат.

## 1. Построение критериев

Рассмотрим некоторые типичные гипотезы, которые на практике приходится проверять, и используемые для них критерии.

### 1.1 Критерии согласия

Для проверки *простых* гипотез (то есть гипотез, состоящих из одного распределения $\mathcal{P}_0 = \{\mathsf{P}_0\}$) используют так называемые *критерии согласия*. Для непрерывных $\mathsf{P}_0$ используют критерий Колмогорова (`scipy.stats.kstest`), критерий Крамера-фон Мизеса-Смирнова (`scipy.stats.cramervonmises`) и аналогичные, для категориальрых данных -- критерий хи-квадрат (`scipy.stats.chisquare`). Все эти критерии используют в качестве статистики расстояние между эмпирическим и теоретическим распределениями: для критерия Колмогорова это $C$-норма разности ф.р., для Крамера-фон Мизеса --- $L_2$-норма, для критерия хи-квадрат --- [дивергения-$\chi^2$](https://en.wikipedia.org/wiki/F-divergence#Common_examples_of_f-divergences). В следующем разделе посмотрим на них более подробно.

### 1.2 Критерии независимости

Другой важной прикладной задачей является выявление зависимости между двумя метриками. Формально говоря, у нас есть выборка из случайных векторов $(X_1, Y_1), \ldots, (X_n, Y_n)$ из распределения $\mathsf{R}$, где $X_i \sim \mathsf{P}$, $Y_i \sim \mathsf{Q}$. Нулевая гипотеза в таком случае будет иметь вид $H_0\colon \mathsf{R} = \mathsf{P} \otimes \mathsf{Q}$. В дискретном случае эта задача не очень сложная, и имеется критерий, чья статистика аналогична статистики критерия хи-квадрат. Если признак $X$ принимает всего $k$ значений, а признак $Y$ --- $m$ значений, то при верности $H_0$ с ростом выборки справедлива сходимость
$$
\chi^2(\mathbf X, \mathbf Y) = \sum_{i=1}^m \sum_{j=1}^k \frac{(\nu_{ij} - \nu_{i\bullet} \nu_{\bullet j} / n)^2}{\nu_{i\bullet} \nu_{\bullet j} / n} \stackrel{d}{\longrightarrow} \chi^2_{(m-1)(k-1)},
$$
где $\nu_{ij}$ --- количество пар с $X = i$ и $Y = j$, $\nu_{i\bullet} = \sum_{j} \nu_{ij}$, $\nu_{\bullet j} = \sum_{i} \nu_{ij}$

*Пример:* В 1892 году в работе Фрэнсиса Гальтона ["Finger Prints"](https://galton.org/books/finger-prints/index.htm) изучалась наследственность типов отпечатков пальца: в виде дуг, петлей и завитков. Для этого бралась выборка из 105 пар братьев и сестёр и сравнивался тип отпечатков пальцев в каждой паре~--- получился набор из 9 чисел, записанных в таблицу сопряжённости: по горизонтали берётся тип отпечатков у первого ребёнка из пары, по вертикали --- второго.

| | Дуги | Петли | Завитки |
|------|------|------|------|
| Дуги | 5 | 12 | 2 |
| Петли | 4 | 42 | 15 |
| Завитки | 1 | 14 | 10 |

Если предположить, что наследственность роли не играет, то зависимости быть не должно. Проверим эту гипотезу на уровне значимости 0.05 с помощью критерия $\chi^2$. Статистику критерия можно посчитать и напрямую, но удобнее занести данные в Python и воспользоваться функцией [`scipy.stats.chi2_contingency`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2\_contingency.html).

```python
fraternal_obs = np.array([
    [5, 12, 2],
    [4, 42, 15],
    [1, 14, 10]
])
print(sps.chi2_contingency(fraternal_obs))
# Или можно посчитать всё напрямую
n = fraternal_obs.sum()
x_num = fraternal_obs.sum(axis=0, keepdims=True)
y_num = fraternal_obs.sum(axis=1, keepdims=True)
stat = ((fraternal_obs - x_num * y_num / n) ** 2 / (x_num * y_num / n)).sum()
pvalue = sps.chi2((3 - 1) * (3 - 1)).sf(stat)
print(f"Ручная проверка: stat - {stat}, pvalue - {pvalue}")
```

Таким образом, гипотеза о независимости данных признаков у братьев и сестёр отклоняется на взятом уровне значимости.

Для непрерывных случайных величин честно проверять независимость -- дело проблематичное, поэтому зачастую используют так называемые *коэффициенты корреляции*, которые при верности нулевой гипотезы околонулевые, но при этом отлавливают не все альтернативы. Например, коэффициент корреляции Пирсона, который просто равен выборочной корреляции выборок:
$$
\widehat{\rho} = \frac{\sum_{i=1}^n (X_i - \overline{X})(Y_i - \overline{Y})}{\sqrt{\sum_{i=1}^n (X_i - \overline{X})^2 \sum_{i=1}^n (Y_i - \overline{Y})^2}}.
$$

В силу предельных теорем данная случайная величина действительно сходится к теоретическому значению корреляции $\mathsf{corr}(X, Y)$, но к сожалению она равна нулю не только в случае независимости, что делает критерии, основанный на подобного рода статистиках, маломощными.

### 1.3 Критерии Вальда

Завершим обсуждение простым способом построения асимптотических критериев посредством асимпт. нормальных оценок. Если таковая у нас имеется:
$$
\sqrt{n} \cdot \frac{\widehat \theta - \theta}{\sigma(\theta)} \stackrel{d}{\to} \mathcal{N}(0, 1),
$$
то, как мы знаем из курса статистики, мы можем построить на её основе доверительный интервал уровня значимости $\gamma$:
$$
\mathsf{P}_{\theta}\left[\theta \in \left(\widehat \theta - \frac{z_{(1+\gamma)/2} \sigma(\widehat \theta)}{\sqrt{n}}; \widehat \theta + \frac{z_{(1+\gamma)/2} \sigma(\widehat \theta)}{\sqrt{n}}\right)\right] \to \gamma.
$$

Если накрытие истинного значения параметра интервалов происходит с вероятностью $\gamma$, то наоборот, непопадание в него произойдёт с вероятностью $1 - \gamma$. Таким образом можно построить критерий уровня значимости $\alpha = 1 - \gamma$ для проверки гипотезы $H_0\colon \theta = \theta_0$: если 
$$
\theta_0 \not\in \left(\widehat \theta - \frac{z_{(1+\gamma)/2} \sigma(\widehat \theta)}{\sqrt{n}}; \widehat \theta + \frac{z_{(1+\gamma)/2} \sigma(\widehat \theta)}{\sqrt{n}}\right)
$$
или, что то же самое,
$$
T(\mathbf X) = \sqrt{n}\frac{|\widehat \theta - \theta_0|}{\sigma(\widehat \theta)} > z_{(1+\gamma)/2},
$$
то гипотеза $H_0$ отвергается.

Так можно строить критерии для проверки, например, $H_0\colon \mu = 0$ по выборке $X_1, \ldots, X_n \sim \mathcal{N}(\mu, \sigma^2)$:
$$
\widehat \theta = \overline{\mathbf X}, \;\;\; T(\mathbf X) = \sqrt{n}\frac{|\overline{\mathbf X}|}{s(\mathbf X)},
$$
где $s^2(\mathbf X)$ --- выборочная дисперсия. Или чуть более сложные гипотезы вида $H_0\colon \mu_1 = \mu_2$ по независимым выборкам $X_1, \ldots, X_n \sim \mathcal{N}(\mu_1, \sigma^2)$, $Y_1, \ldots, Y_n \sim \mathcal{N}(\mu_2, \sigma^2)$:
$$
\widehat \theta = \overline{\mathbf X} - \overline{\mathbf Y}, \;\;\;T(\mathbf X) = \sqrt{n}\frac{|\overline{\mathbf X} - \overline{\mathbf Y}|}{\sqrt{s^2(\mathbf X) + s^2(\mathbf Y)}}.
$$

Однако в случае нормальных выборок распределение статистик выше можно уточнить и получить точные, а не асимптотические критерии. Вспомним, что по теореме Фишера $\overline{\mathbf X}$ не зависит от $s^2(\mathbf X)$, причём $ns^2(\mathbf X) / \sigma^2 \sim \chi^2_{n-1}$. Отношение независимых нормальной с.в. и корня из хи-квадрат с параметром $k$ распределён как $T_{k}$:
$$
\xi \sim \mathcal{N}(0, 1), \eta \sim \chi^2(k) \Longrightarrow \frac{\xi}{\sqrt{\eta / k}} \stackrel{def}{\sim} T_{k}
$$
поэтому для $X_1, \ldots, X_n \sim \mathcal{N}(\mu, \sigma^2)$
$$
\sqrt{n-1} \frac{\overline{X} - \mu}{s(\mathbf X)} \sim T_{n-1},
$$
а для независимых $X_1, \ldots, X_n \sim \mathcal{N}(\mu_1, \sigma^2)$, $Y_1, \ldots, Y_m \sim \mathcal{N}(\mu_2, \sigma^2)$
$$
\sqrt{\frac{nm}{n+m}} \frac{\overline{X} - \overline{Y} - (\mu_1 - \mu_2)}{\sqrt{\frac{(n-1)s^2(\mathbf X) + (m-1)s^2(\mathbf Y)}{n+m-2}}} \sim T_{n+m-2}.
$$

В следующем практикуме вы немного поработаете с этим критерием и сравните его с некоторыми другими критериями для проверки равенства средних (в том числе с критерием Вальда).

## 2. Гороховые законы

### 2.0 *История*

В XIX веке австрийский монах и по совместительству биолог Грегор Мендель представил миру результаты своих опытов по скрещиванию гороха с различными визуальными признаками, на основании которых выдвинул гипотезы о наследовании признаков от родительских особей к их потомках. Широкому читателю они известны под названием ["Законы Менделя"](https://ru.wikipedia.org/wiki/Законы_Менделя). Мы не будем глубоко погружаться в эту тему, но рассмотрим второй из них (закон расщепления), который утверждает, что 

> При скрещивании двух гетерозиготных особей (то есть обладающих двумя разными аллелями, отвечающими одному признаку) среди их потомков наблюдается расщепление в их фенотипе в отношении 3:1

Расшифровка вышесказанного такова. Пусть у двух размножающихся особей два различных аллеля гена (Aa), отвечающих какому-то признаку (например, цвет гороха, его форма и тд), обычно их называют доминантным (A) и рецессивным (a), так как в такой ситуации первый аллель подавляет второй и визуально (по фенотипу) особь не отличается от особи с обеми доминантными аллелями. При зарождении гаметы потомка на роль двух аллелей соответствующего гена претендует один из аллелей от каждого родителя. Закон утверждает, что выжимаемость гаметы (а значит и появление потомка) не зависит от итоговой комбинации аллелей, поэтому в итоге среди потомства где-то 1/4 особей будет иметь генотип AA, 2/4 -- генотип Aa, а 1/4 --- генотип aa, а значит по фенотипу 3/4 потомства будет иметь доминантный признак, а 1/4 -- рецессивный.

Если вы ничего не поняли из написанного выше, то попробуйте воспринять второй закон Менделя статистически. Потомство представляет собой выборку из распределения Бернулли:
$$
X_1, \ldots, X_n \sim \mathrm{Bern}(p),
$$
где $X_i$ равен 1, если $i$-ый потомок обладает доминантный признаком, и 0, если рецессивным. Утверждение выше можно трактовать как статистическую гипотезу $H_0\colon p = 3/4$.

Советская школа товарища Т. Д. Лысенко не принимала эту точку зрения и считала, что неравная жизнеспособность набора аллелей имеет место быть и играет существенную роль, чтобы поставить под сомнение распределение, описываемое гипотезой $H_0$. При взгляде на эту картинку может возникнуть сомнения в верности $H_0$, так как нередки случаи большого уклонения от теоретического значения. Или всё же редки?

### 2.1 Наивная проверка

Для начала объединим все выборки в одну, сложив количество потомков для каждой аллели. Если выборки независимы, то итогое распределение останется биномиальным с параметром $p=3/4$ при верности $H_0$.

Гипотезу о том, что некоторая категориальная модель имеет определённые вероятности, можно проверить с помощью *критерия хи-квадрат*. Напомним его вид в общем случае. Пусть выборка $X_1, \ldots, X_n$ такова, что при нулевой гипотезе $\mathsf{P}(X_1 = i) = p_i$ для $i = 1, \ldots, k$, где $p_i > 0$, $\sum p_i = 1$. Теорема Пирсона утверждает, что при верности нулевой гипотезы
$$
\chi^2(\mathbf X) = \sum_{i=1}^k \frac{(\nu_i - np_i)^2}{np_i} \stackrel{d}{\to} \chi^2_{k-1},
$$
где $\nu_i = \sum I(X_j = i)$ --- количество выпаданий класса $i$ в выборке. В нашем случае $k=2$ и $p_1 = 3/4$. 

Для проверки данной гипотезы воспользуемся функцией [`scipy.stats.chisquare(f_obs, f_exp)`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chisquare.html#scipy.stats.chisquare), где `f_obs` --- наблюдаемые частоты, а `f_exp` --- теоретические, то есть $f\_exp_i = n \cdot p_i$, где $n$ --- размер выборки, $p_i$ --- вероятность выпадания $i$-ого класса при верности нулевой гипотезы.

```python
obs = table.sum(axis=0)
print(f"A - {obs[0]}\na - {obs[1]}")
n = obs.sum()
exp = n * np.array([3/4, 1/4])
sps.chisquare(obs, exp)
```

На уровне значимости 0.05 гипотеза $H_0$ не отвергается. Но насколько корректна данная процедура? Вдруг, например, вероятность $p$ в каждой конкретной выборке какая-то своя, а итоговое соответствие нулевой гипотезе --- следствие тупого усреднения по всем выборкам? Такого рода ситуации вполне возможны, и критерий хи-квадрат против них крайне немощен.

Напишем простенькую функцию, которая генерирует несколько выборок из биномиального распределения, чей параметр случаен (для простоты размер каждой выборки будет всегда одним и тем же), и применим полученные синтетические данные к нашему методу.

```python
def generate_samples(n_samples, min_p=0.6, max_p=0.9, min_n=30, max_n=30):
    p = sps.uniform(loc=min_p, scale=max_p-min_p).rvs(size=n_samples)
    n = sps.randint(min_n, max_n+1).rvs(size=n_samples)
    m = sps.binom(n, p).rvs(size=n_samples)
    obs = np.stack([m, n - m], axis=1)
    return obs
```

```python
samples = generate_samples(20)
obs = samples.sum(axis=0)
print(f"A - {obs[0]}\na - {obs[1]}")
n = obs.sum()
exp = n * np.array([3/4, 1/4])
sps.chisquare(obs, exp)
```

Как можно видеть, в данной ситуации, которая отличается от гипотезы $H_0$, критерий также всё устраивает.

Данное обстоятельство не позволяет небрежно "класть все яйца в одну корзину" и вынуждает проверять гипотезу для каждой выборки в отдельности. Но что с этим делать дальше? Ведь в каких-то случаях гипотеза будет отвергаться, а в каких-то --- нет.

### 2.2 Анализ pvalue

Итак, вот мы применили критерий для каждой выборки и посчитали соответствующие pvalue. 

```python
sample_sizes = table.sum(axis=-1, keepdims=True)
exp = sample_sizes * np.array([3/4, 1/4])
pvalues = sps.chisquare(table, exp, axis=-1).pvalue
print("Некоторые значения pvalue:", pvalues[:5])
```

Вспомним определение pvalue --- это *вероятность того, что значение статистики критерия окажется больше того, что мы наблюдаем на самом деле, при верности $H_0$*, что можно переписать как
$$
pvalue(\mathbf x) = \mathsf{P}_{X \sim H_0}(T(\mathbf X) \ge T(\mathbf x)) = 1 - F_{T(\mathbf X)}(T(\mathbf X)),
$$
где $\mathbf X$ --- выборка, $T(\mathbf X)$ --- статистика критерия. Заметим, что случайная величина $F_{T(\mathbf X)}(T(\mathbf X))$ распределена равномерно на отрезке $[0; 1]$ (известная задачка из курса по теории вероятностей), а значит и $pvalue(\mathbf X)$ тоже.

Таким образом, если $H_0$ верна, то выборка из посчитанных нами $pvalue$ распределена как $\mathrm{U}[0; 1]$. Это сводит задачу к проверке гипотезы о том, что выборка пришла из какого-то конкретного распределения. Её можно решить с помощью, например, критерия Колмогорова, который основан на статистике
$$
D_n = \sup_{x \in \mathbb{R}} |\widehat{F}_n(x) - F_0(x)|,
$$

где $\widehat{F}_n$ --- ECDF, построенная по выборке $X_1, \ldots, X_n$. Теорема Колмогорова гласит, что при верности нулевой гипотезы $H_0\colon X_i \sim F_0$ имеется сходимость $\sqrt{n} D_n \stackrel{d}{\to} \xi$, где $\xi$ имеет известное распределение.

Считать статистику самому, да ещё и сверять её с квантилями какого-то противного распределения -- себя не уважать, так что воспользуемся функцией [`scipy.stats.kstest(rvs, cdf)`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html), где `rvs` --- выборка, а `cdf` --- предполагаемая функция распределения $F_0$. В нашем случае `cdf = sps.uniform.cdf`.

```python
cdf = sps.uniform.cdf
pvalue = sps.kstest(pvalues, cdf).pvalue
print(f"Pvalue для KS-теста: {pvalue:0.3f}")
```

Для наглядности визуализируем CDF и ECDF. 

```python
grid = np.linspace(-0.1, 1.1, 1000)
ecdf = ECDF(pvalues)

plt.plot(grid, ecdf(grid), color='red', label='ECDF по выборке pvalue')
plt.plot(grid, sps.uniform.cdf(grid), color='green', label='Распределение при $H_0$')

plt.title("Сравнение эмпирической и теоретической ф.р.")
plt.legend()
plt.show()
```

Также для примера можно посмотреть на результат работы [критерия Крамера-фон Мизеса-Смирнова](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.cramervonmises.html#reed8bce1e242-1), статистика которого основана уже не на $C$-норме, а на интеграле разности функций распределения.

```python
cdf = sps.uniform.cdf
cvms_pvalue = sps.cramervonmises(pvalues, cdf).pvalue
print(f"Pvalue для теста Крамера-фон Мизеса-Смирнова: {cvms_pvalue:0.3f}")
```

Для уверенности применим новый способ на тех же синтетических данных и посмотрим, как распределены pvalue в этом и предыдущем способах.

```python
straightforward_pvalues = []
updated_pvalues = []
for _ in tqdm(range(10000)):
    samples = generate_samples(30)

    # Первый способ
    obs = samples.sum(axis=0)
    n = obs.sum()
    exp = n * np.array([3/4, 1/4])
    straightforward_pvalues.append(sps.chisquare(obs, exp).pvalue)

    # Второй способ
    sample_sizes = samples.sum(axis=-1, keepdims=True)
    exp = sample_sizes * np.array([3/4, 1/4])
    pvalues = sps.chisquare(samples, exp, axis=-1).pvalue
    updated_pvalues.append(sps.kstest(pvalues, sps.uniform.cdf).pvalue)
```

```python
plt.hist(straightforward_pvalues, alpha=0.8, density=True, label='Сваливаем выборки в одну кучу')
plt.hist(updated_pvalues, alpha=0.8, density=True, label='Применяем KS-тест к pvalue выборок')

plt.title("Распределения pvalue для разных способов проверки $H_0$")
plt.legend()
plt.show()
```

Наш новый подход крайне хорошо справляется с такого рода альтернативами, в то время как топорный метод весьма посредственен.

### 2.3 Подход Колмогорова

В статье "Об одном новом подтверждении законов Менделя" А. Н. Колмогоров также разбирает полученные Н. И. Ермолаевой результаты, но делает это немного иначе (действительно, считать в 1940 году 200 штук pvalue было бы неудобно): он пользуется нормальным приближением биномиального распределения. По ЦПТ распределение случайной величины $\frac{\xi - np}{\sqrt{npq}}$, где $\xi \sim \mathrm{Bin}(n, p)$, можно приблизить стандартным нормальным распределением. Посредством этого факта задача сводится к проверки нормальности некоторого распределения. Сделать это можно с помощью того же критерия Колмогорова, что он и сделал в своей заметке. Повторим процедуру.

```python
sample_sizes = table.sum(axis=-1)
p = 0.75
normed_table = (table[:, 0] - sample_sizes * p) / np.sqrt(sample_sizes * p * (1 - p))
pvalue = sps.kstest(normed_table, sps.norm.cdf).pvalue

grid = np.linspace(-3.5, 3.5, 1000)
ecdf = ECDF(normed_table)
plt.figure(figsize=(10, 6))
plt.plot(grid, ecdf(grid), color='red', label='ECDF по нормализованной выборке')
plt.plot(grid, sps.norm.cdf(grid), color='green', label='Приблизительное распределение при $H_0$')

plt.title(f"Pvalue для KS-теста: {pvalue:0.4f}")
plt.legend()
plt.show()
```

Приближение на первый взгляд кажется довольно точным, да и KS-тест всё устраивает. С другой стороны, если присмотреться, то заметен довольно тяжёлый левый хвост, хотя у нормального распределения он очень лёгок.

Это может быть проявлением маломощности критерия Колмогорова: он достаточно плохо обнаруживает отклонения по краям. Для случая нормальности часто применяют так называемый критерий Шапиро-Уилка, который признаётся одним из мощнейших критериев для проверки нормальности (обратите внимание, он проверяет не просто равенство распределения какому-то конкретному нормальному, как KS-тест, а *нормальность в принципе*). Не будем вдаваться в подробности его устройства и воспользуемся готовой реализацией [`scipy.stats.shapiro`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html). Также для наглядности построим QQ-plot, который помогал нам ранее в визуальной проверке нормальности.

```python
shapiro_pvalue = sps.shapiro(normed_table).pvalue
sps.probplot(normed_table, dist=sps.norm, plot=plt)
plt.title(f"Pvalue критерия Шапиро-Уилка - {shapiro_pvalue:0.8f}")
plt.show()
```

Теперь уже отклонения от нормальности более ярковыраженные, да и критерий Шапиро-Уилка резко отклоняет гипотезу о нормальности. Моожет быть, гипотеза и вправду неверна, и мы зря потешались над Ермолаевой?

Давайте вспомним, что размер у каждой выборки свой, и иногда он может быть довольно маленьким.

```python
plt.hist(sample_sizes, bins=20)
plt.title("Размеры потомств")
plt.show()
```

Как можно видеть, среди данных есть потомства размера 10 и меньше. Вряд ли статистики, построенные по таким выборкам выше, действительно хорошо приближаются нормальным распределением, не так ли? В частности, это касается критерия хи-квадрат, который вообще-то тоже основан на ЦПТ, а значит, предельное распределение является состоятельным приближением настоящего лишь для достаточно большой выборки. Из-за подобных 

Это приводит нас к необходимости **валидации критерия** -- проверки того, что используемый критерий обладает требуемый уровнем значимости. Для этого достаточно уметь генерировать данные из нулевой гипотезы, тогда по распределению pvalue мы поймём адекватность критерия: в идеале оно должно быть равномерным, иначе --- скошенным в сторону нуля.

Для проверки будет сэмплировать данные, в которых размер потомства будет случайным, в том числе может быть довольно малым (остановимся на пяти).

```python
shapiro_pvalues = []
ks_pvalues = []
n_iter = 50000
for _ in tqdm(range(n_iter)):
    # Сэмплируем данные, которые удовлетворяют нулевой гипотезе
    samples = generate_samples(100, min_p=0.75, max_p=0.75, min_n=5, max_n=30)

    # Сначала проверим критерий на основе Шапиро-Уилка
    p = 0.75
    sample_sizes = samples.sum(axis=-1)
    normed_table = (samples[:, 0] - sample_sizes * p) / np.sqrt(sample_sizes * p * (1 - p))
    shapiro_pvalues.append(sps.shapiro(normed_table).pvalue)

    # Теперь проверим критерий на основе хи-квадрат и KS-теста
    exp = sample_sizes[:, None] * np.array([3/4, 1/4])
    pvalues = sps.chisquare(samples, exp, axis=-1).pvalue
    ks_pvalues.append(sps.kstest(pvalues, sps.uniform.cdf).pvalue)
```

Заодно давайте посчитаем ДИ для ошибки I рода, если в качестве порога для отклонения гипотезы брать уровень значимости 0.05. Если критерий корректен, то взятый уровень значимости должен быть в интервале (или хотя бы лежать около него). Если же интервал лежит правее взятого уровня значимости, то истинная ошибка I рода у такого критерия гораздо больше планируемого, что не есть хорошо.

```python
shapiro_confint = proportion_confint((np.array(shapiro_pvalues) < 0.05).sum(), n_iter)
ks_confint = proportion_confint((np.array(ks_pvalues) < 0.05).sum(), n_iter)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(shapiro_pvalues, bins=20)
plt.title(f"Критерий на основе Шапиро-Уилка\nДИ ошибки I рода - ({shapiro_confint[0]:0.3f}, {shapiro_confint[1]:0.3f})")

plt.subplot(1, 2, 2)
plt.hist(ks_pvalues, bins=20)
plt.title(f"Критерий на основе хи-квадрат и KS-теста\nДИ ошибки I рода - ({ks_confint[0]:0.3f}, {ks_confint[1]:0.3f})")
plt.show()
```

Оказывается, критерии и вправду завышают ошибку I рода. Процедура по KS-тесту ещё куда не шло, а вот с проверкой с помощью Шапиро-Уилка всё очень плохо: настоящая вероятность отвергнуть истинную нулевую гипотезу больше планируемой в 4 раза! С такими выводами доверять отклонению нулевой гипотезы в нашем случае довольно рискованно. 

Посмотрим, что будет с pvalue, если потомства будут достаточно большого размера (для определённости возьмём минимум 20).

```python
shapiro_pvalues = []
ks_pvalues = []
n_iter = 50000
for _ in tqdm(range(n_iter)):
    # Сэмплируем данные, которые удовлетворяют нулевой гипотезе
    samples = generate_samples(100, min_p=0.75, max_p=0.75, min_n=20, max_n=40)

    # Сначала проверим критерий на основе Шапиро-Уилка
    p = 0.75
    sample_sizes = samples.sum(axis=-1)
    normed_table = (samples[:, 0] - sample_sizes * p) / np.sqrt(sample_sizes * p * (1 - p))
    shapiro_pvalues.append(sps.shapiro(normed_table).pvalue)

    # Теперь проверим критерий на основе хи-квадрат и KS-теста
    exp = sample_sizes[:, None] * np.array([3/4, 1/4])
    pvalues = sps.chisquare(samples, exp, axis=-1).pvalue
    ks_pvalues.append(sps.kstest(pvalues, sps.uniform.cdf).pvalue)
```

```python
shapiro_confint = proportion_confint((np.array(shapiro_pvalues) < 0.05).sum(), n_iter)
ks_confint = proportion_confint((np.array(ks_pvalues) < 0.05).sum(), n_iter)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(shapiro_pvalues, bins=20)
plt.title(f"Критерий на основе Шапиро-Уилка\nДИ ошибки I рода - ({shapiro_confint[0]:0.3f}, {shapiro_confint[1]:0.3f})")

plt.subplot(1, 2, 2)
plt.hist(ks_pvalues, bins=20)
plt.title(f"Критерий на основе хи-квадрат и KS-теста\nДИ ошибки I рода - ({ks_confint[0]:0.3f}, {ks_confint[1]:0.3f})")
plt.show()
```

Не лучшая картинка, критерий на основе Шапиро всё ещё некорректен, но стало гораздо лучше, да и ошибка I рода у второго критерия стала ближе к требуемой.

Давайте проверим, что станет с результатами применения критерия в случае, если мы будем отбрасывать малые (размера <20) потомства.

```python
sample_sizes = table.sum(axis=-1)

mask = (sample_sizes >= 20)
print(f"Осталось выборок: {mask.sum()}")
upd_sizes = sample_sizes[mask]
upd_table = table[mask]

p = 0.75
norm_sample = (upd_table[:, 0] - p * upd_sizes) / np.sqrt(upd_sizes * p * (1 - p))
sps.shapiro(norm_sample)
```

Теперь уже критерий Шапиро-уилка не отвегает нулевую гипотезу. Если внимательно следить за корректностью применения критериев, то можно избежать небрежных ошибок I рода.

Однако важно понимать: **не стоит воспринимать всё вышесказанное как 100%-ое доказательство того, что закон Менделя работает**. Мы всего лишь показали с помощью статистики, что нет каких-то серьёзных оснований для отклонения этого закона и необъяснимое (кроме как неверностью закона расщепления) отклонение от истинной доли 3/4 вполне себе укладываются в границы случайности.

Таким образом, pvalue можно использовать не только для непосредственно проверки гипотез (aka сравнением pvalue со взятым уровнем значимости), но и для
- Валидации критериев (проверки того, что распределение статистики совпадает с теоретическим при верности нулевой гипотезы);
- Сравнении чувствительности критериев (насколько хорошо они отклоняют альтернативные гипотезы).
В этих задачах также помогает критерии согласия (например, критерий Колмогорова), который показывает степень расхождения предполагаемого распределения и наблюдаемоего.
