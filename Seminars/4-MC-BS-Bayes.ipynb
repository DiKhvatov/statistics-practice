### Задача 1. Бутстрапирование

Когда приходится работать с точечными оценками, для которых не известно распределение встаёт вопрос о том, чему равна дисперсия оценки, риск и квантили. С одной стороны, мы умеем восстанавливать распределение оценки при помощи KDE, которое строится по выборке оценок. Но это в случае, когда мы имеем неограниченный доступ к исходному распределению иксов. На практике сбор данных представляет собой отдельную трудоёмкую задачу (требующую, к примеру, проведения соц. опроса или контрольных замеров), так что понижение размера выборки с $N$ до $\sqrt{N}$ (а именно такого размера будут выборки и будет их количество, если всего в наличии есть $N$ элементов) оказывается недопустимо. При этом при публикации оценённых параметров (к примеру, вероятности заболевания гриппом осенью) вопрос о том, с какой точностью эти параметры замерены (какова дисперсия предложенной оценки) остаётся очень важным.

В общем случае ставится следующий вопрос. $X_i \sim P_{\theta}$. Дана какая-то интегрируемая функция $f: R \rightarrow \mathbb{R}$ и оценка $\theta^*$. Требуется посчитать $\int\limits_{\mathbb{R}} f(\theta^*) dP_{\theta}$. Заметим, что задача о поиске мат. ожидания, дисперсии и квантилей оценки сводятся к одной этой задаче.

Брать интеграл точно оказывается невозможным, так как $P_{\theta}$ неизвестно. По УЗБЧ можно приблизить этот интеграл как среднее $\overline{f(\theta^*)}$, однако для этого нужно несколько выборок.

Приближение интеграла при помощи среднего носит название **метода Монте-Карло**

Важным инструментом для борьбы с этой проблемой оказывается теорема Гливенко-Кантелли, которая говорит о том, что эмпирическая функция распределения $\hat{F}(x)$ сходится равномерно почти наверное к истинной функции распределения $F(x)$, а значит и эмпирическая мера в каком-то смысле (детали оставляем на откуп лектору) сходится к истинной мере. А значит, и мат. ожидание по эмпирической мере сходится к мат. ожиданию по исходной мере.



Тогда интеграл $\int\limits_{\mathbb{R}} f(\theta^*) dP_{\theta}$ можно приближать как $\int\limits_{\mathbb{R}} f(\theta^*) d\hat{P}_{\theta}$. Такой интеграл представляет собой сложную сумму из экспоненциального числа слагаемых, и брать его напрямую тоже не представляется возможным. Но уже его можно брать при помощи Монте-Карло, сэмплируя выборки из $\hat{P}$. Именно это и называется бутстрапированием.

Для доказательства того, что бутстрепное среднее сходится к мат. ожиданию по истинной мере требуется применить две сходимости: сходимость бутстрепного среднего к мат. ожиданию по эмпирической мере $\hat{P}$ и сходимость этого мат. ожидания к мат. ожиданию по исходной мере $P$. По этой причине хорошим тоном считается брать количество бутстрепных выборок равным количеству элементов каждой бутстрепной выборки, и равным количеству элементов исходной выборки.

Теоретически эта рекомендация обоснована слабо, так как скорости сходимости в обоих случаях разные.

```python
N = 1000 # sample size
mu = 0.3 # expectation
X = sps.bernoulli.rvs(p=0.3, size=N)
mu_hat = np.mean(X)
print(mu_hat)
```

В типичном для практике случае мы не знаем, чему равен параметр и задаёмся вопросом о дисперсии оценки и о том, в каком интервале оценка находится с высокой вероятностью (позже мы назовём это **доверительным интервалом**).

Вопрос: почему сэмплировать надо с возвращением?

```python
bootstrap = np.random.randint(0, N, (N, N))
mu_hat = X[bootstrap].mean(-1)
print(mu_hat.mean(), mu_hat.var())
```

```python
X = sps.bernoulli.rvs(p=0.3, size=(N, N))
mu_hat = X.mean(-1)
print(mu_hat.var())
```

### Байесовское оценивание

Байесовское оценивание даёт ультиматимативный способ построения оценок без какой-либо потери в информации. Теперь мы оцениваем параметры распределения не точками и не отрезками, а распределениями!

Недостатки такого подхода очевиден: распределение параметра нужно как-то эффективно хранить (а это значит, что его тоже нужно параметризовать, но мы не имеем гарантии, что наша параметризация вообще имеет смысл в приложении к реальным законам природы) и как-то считать по формуле Байеса.

Но преимущества выглядят очень полезными:
- Мы знаем не только одно значение параметра, а имеем некоторое знание о поведении этого параметра

- Мы можем корректировать оценку некоторым априорным знанием, чтобы не подгоняться под практику и уточнять оценку в зависимости от требований практики

- В случае, если модель носит предсказтельный характер (пояснено ниже), то мы применяем взвешенное голосование вместо применения одной модели

- Знания о параметре можно уточнять с приходом новых данных в онлайне

Выделим два важных класса моделей в статистике: генеративные и дискриминантные.

С генеративными моделями мы уже встречались: $X \sim P(\theta)$. В рамках этой модели изучается распределение из которого приходят данные. Название означает, что изучение параметров позволяет напрямую генерировать новые данные.

Дискриминантные модели выглядят иначе: $Y \sim P(X, \theta)$, где $X$ --- случайная величина из какого-то распределения, о котором неизвестно вообще ничего (даже семейства). Например, (X, Y) это банковские данные, где $X$ --- какие-то измеренные параметры (заработок, наличие образования, количество сменённых мест работы, etc.), а $Y$ --- индикатор того, вернул ли человек кредит. Нас не очень интересует, как сэмплируется $X$, но нас интересует, как зависит $Y$ от $X$, то есть условное распределение $P(Y | X)$. Именно оно полагается принадлежащим какому-то параметрическому семейству с параметрами $\theta$. Теперь становится понятным польза байесовского оценивания: если мы знаем распределение параметра $P(\theta | X, Y)$, то мы можем узнать распределение $Y$ при условии данных $X$, то есть по новому человеку понять, какова вероятность того, что кредит он нам вернёт.

Выкладки далее будут делаться для дискриминантных моделей. Заметим, что для генеративных моделей они получаются аналогичным образом.

Итак, для применения формулы Байеса $P(\theta | X, Y) = \frac{P(Y | \theta, X) P(\theta | X)}{P(Y|X)}$ нам нужно знать три величины. $P(Y | \theta, X)$ нам известно --- это наша модель из параметрического семейства, она же правдоподобие. Величина в знаменателе напрямую зависит от величины в числителе: $P(Y|X) = \int_{\mathbb{R}} P(Y | \theta, X) P(\theta|X) d\theta$. Остаётся понять, что такое $P(\theta | X)$. Это некоторое априорное знание о параметре и оно подбирается аналитиками в зависимости от средних показателей или других знаний о природе. Часто полагается $P(\theta | X) = P(\theta)$, то есть рассматриваются модели, в которых параметр одинаково распределён для любых данных. В случае кредитного скоринга это может быть распределение вероятности вернуть кредит. Например, если в среднем кредит возвращают, то оно может быть таким:

```python
grid = np.arange(0, 1, 0.001)
plt.plot(grid, sps.beta.pdf(grid, 10, 1, loc=0, scale=1), label=r'$\theta$')
plt.title('Вероятность возврата кредита')
plt.legend()
plt.show()
```

А если в среднем не возвращают, то таким:

```python
grid = np.arange(0, 1, 0.001)
plt.plot(grid, sps.beta.pdf(grid, 1, 20, loc=0, scale=1), label=r'$\theta$')
plt.title('Вероятность возврата кредита')
plt.legend()
plt.show()
```

А если аналитик не готов брать на себя ответственность говорить что-то об обществе, то таким:

```python
grid = np.arange(0, 1, 0.001)
plt.plot(grid, sps.beta.pdf(grid, 5, 5, loc=0, scale=1), label=r'$\theta$')
plt.title('Вероятность возврата кредита')
plt.legend()
plt.show()
```

Это априорное распределение транслирует следующую мысль: мы будем считать, что люди равновероятно отдают или не отдают кредит ($\theta = 0.5$), но наша уверенность достаточно слабая. Может так оказаться, что $\theta = 0.6$ или $0.7$.

А если аналитик уверен в себе, то

```python
grid = np.arange(0, 1, 0.001)
plt.plot(grid, sps.beta.pdf(grid, 100, 100, loc=0, scale=1), label=r'$\theta$')
plt.title('Вероятность возврата кредита')
plt.legend()
plt.show()
```

Итак, априорное распределение мы подбираем сами и оно задаёт какие-то исходные знания о законах природы.

 $P(\theta | X, Y) = \frac{P(Y | \theta, X) P(\theta | X)}{P(Y|X)}$. Как мы поняли, в знаменателе находится интеграл. Брать его оказывается трудоёмко на практике, однако он представляет собой какую-то константу от тета независящую, так что для начала попробуем понять, что можно полезного узнать, не считая его.

 А узнать можно, например, максимум.

Подход оценивания параметра максимумом апостериорного распределения получил название Байес для бедных (Poor man's Bayes) и очень похож на оценивание максимумом правдоподобия. Возьмём минус логарифм от плотности и будем считать его минимум, как мы привыкли.

$\theta^* = argmax(-log(P(Y | \theta, X) + (-log P(\theta))))$. Если априорное распределение взять нормальным $N(0, \lambda)$, то второе слагаемое имеет вид $\frac{\lambda}{2}\theta^2$ и оно добавляется к правдоподобию.

Будем считать, что мы работаем с генеративной моделью и оцениванием вероятность возврата кредита не зная ничего о людях

Ищется минимум функции $log(\theta)(\sum X_i + \alpha - 1) + log(1-\theta)(n - \sum X_i + \beta - 1)$

Проверьте, что он достигается в точке $\theta^* = \frac{\sum X_i + \alpha - 1}{n + \alpha - 1 + \beta - 1}$, $\alpha, \beta$ --- параметры априорного распределения

Таким образом, если ошибка для нас имеет разные последствия (к примеру, выдать кредит тому, кто не вернёт строго опаснее, чем не выдать кредит тому, кто вернёт), то можно устроить априорное распределение с пиком в более "безопасной" области параметра. Таким образом, чтобы "убедить" нас дать кредит (максимум апостериорной плотности был ближе к единице) среднее должно быть больше.

**Общий принцип:** априорное распределение отражает степень уверенности нас в том, чему равен параметр. Чем оно вырожденнее, тем больше нужно данных и более высокое правдоподобие, чтобы его изменить.


А если я не хочу быть бедным?

Интеграл в знаменателе во многих случаях можно не вычислять, а узнать исходя из теоретических соображений.

Если известно, что плотность имеет вид $p(\theta) = C f(\theta)$, где $f(x)$ --- интегрируемая функция, то можно сразу узнать параметр $C = \frac{1}{\int f(x) dx}$, так как плотность интегрируется в единицу. Стало быть, чтобы узнать нормировку в знаменателе нужно точно понять, какому параметрическому семейству принадлежит плотность и как меняются его параметры после вариационного байесовского вывода. Тогда нормировочная константа будет известна.

**Определение:** для распределения $P(\theta)$ сопряжённым называется такое распределение $P(Y | X, \theta)$, что распределение $P(\theta | X, Y)$ лежит в таком же параметрическом семействе, что и $P(\theta)$.

Известно, что для распределений из экспоненциального семейства априорное всегда существует. Для примера вернёмся к кредитному скорингу:

Будем писать $p(x) \sim f(x)$ и иметь ввиду, что функции $p(x)$ и $f(x)$ отличаются на константу, причём $p(x)$ это плотность.

$P(\theta) \sim \theta^{\alpha - 1}(1 - \theta)^{\beta-1}$

$P(\theta | X) \sim \theta^{\sum X_i}(1 - \theta)^{n - \sum X_i}\theta^{\alpha - 1}(1 - \theta)^{\beta-1}$

Отсюда легко узнать параметры апостериорного распределения: $\alpha += \sum X_i$, $\beta += n - \sum X_i$. Посмотрим, как меняются апостериорные плотности у разных менеджеров.

```python
grid = np.arange(0, 1, 0.001)
plt.rcParams["figure.figsize"] = (15,7)
plt.plot(grid, sps.beta.pdf(grid, X.sum() + 10, N - X.sum() + 1, loc=0, scale=1), label='\"кредит отдают\"')
plt.plot(grid, sps.beta.pdf(grid, X.sum() + 1, N - X.sum() + 20, loc=0, scale=1), label='\"кредит не отдают\"')
plt.plot(grid, sps.beta.pdf(grid, X.sum() + 5, N - X.sum() + 5, loc=0, scale=1), label='\"я не знаю, отдают ли кредит\"')
plt.plot(grid, sps.beta.pdf(grid, X.sum() + 100, N - X.sum() + 100, loc=0, scale=1), label='\"я знаю, что кредит равновероятно отдают и не отдают\"')
plt.title('Вероятность возврата кредита')
plt.legend()
plt.xlim(0.2, 0.4)
plt.show()
```

Во всех случаях мы будем находить параметр при помощи вариационного байесовского вывода. Но как понять, какая модель лучше описывает данные?

Использовать максимальное правдоподобие тут нельзя, так как оно при добавлении новых параметров только растёт (докажите, почему).

На помощь тут приходит (внезапно) философия, а именно критерий фальсифицируемости научного знания Карла Поппера, который гласит "научная теория не может быть принципиально неопровержимой".

Это значит, что модель, которая может обосновать любые данные правильным подбором параметра менее приоритетна, чем модель, которая может обосновать наблюдаемые данные лучше, чем ненаблюдаемые. Так как интеграл от любой плотности равен единице, то в модели, которая имеет высокий максимум для многих датасетов вероятность самих датасетов меньше, чем в модели, которая имеет высокий максимум для *некоторых* датасетов.

Проще говоря, лучше та модель, в которой в среднем вероятность встретить конкретно наши данные больше.

Оптимизируемая величина это $P(Y | X)$ --- знаменатель дроби в формуле Байеса и называется она **обоснованностью**. В нашем случае $Y$ это результат экзамена, а $X$ это два параметра --- настроение экзаменатора и подготовленность студента.

В качестве априорного распределения берём $Beta(1, 1)$

Первая модель имеет один параметр --- вероятность получить уд на экзамене. Посчитаем обоснованность.

$E_1 = \int_{\theta} P(Y|X, \theta)$, где $P(Y | X, \theta)$ это произведение вероятностей получить уд по всем 4 категориям.

```python
from scipy.special import binom, comb, beta
E1 = comb(151, 19)*comb(9, 0)*comb(63, 11)*comb(103, 5)*beta(37, 293)
print(E1)
```

Вторая модель имеет два параметра --- вероятность получить уд на экзамене у доброго и у злого преподавателя вне зависимости от подготовленности. Посчитаем обоснованность.

$E_1 = \int_{\theta_1, \theta_2} P(Y|X, \theta_1)P(Y|X, \theta_2)$, где $P(Y | X, \theta_i)$ это произведение вероятностей получить уд по категориям преподавателей.

```python
from scipy.special import binom, comb, beta
E2 = comb(151, 19)*comb(9, 0)*comb(63, 11)*comb(103, 5)*beta(31, 185)*beta(7, 107)
print(E2)
```

Третья модель аналогична предыдущей, но теперь параметры отвечают за подготовленность и правдоподобие умножается по категориям студентов

```python
from scipy.special import binom, comb, beta
E3 = comb(151, 19)*comb(9, 0)*comb(63, 11)*comb(103, 5)*beta(20, 141)*beta(18, 149)
print(E3)
```

Ну, и наконец модель с 4 параметрами где все группы интегрируются отдельно.

```python
from scipy.special import binom, comb, beta
E4 = comb(151, 19)*comb(9, 0)*comb(63, 11)*comb(103, 5)*beta(152, 20)*beta(10, 1)*beta(64, 12)*beta(104, 6)
print(E4)
```

Теперь становится понятно, чем байесовская оценка плотностью лучше точечной. Пусть к нам пришёл на вход вектор признаков (настроение преподавателя, подготовленность студента) и мы хотим узнать результат экзамена.

Считать его надо так $P(Y | X) = \int_{\theta} P(Y | X, \theta) P(\theta)$. Если мы знаем распределение $P(\theta)$, то эту плотноть можно узнать точно и вот ответ. Если нет, то этот интеграл можно приблизить по его значению в точке $\theta^*$ --- оценке байеса для бедных. Собственно, что в примитивном машинном обучении и делают --- подставляют найденный набор параметров в модель и вычисляют наиболее правдоподобный $Y$. Но, согласимся, что мода распределения несёт меньше информации, чем само распределение
