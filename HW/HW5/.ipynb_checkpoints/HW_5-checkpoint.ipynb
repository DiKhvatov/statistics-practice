## Распределение Стьюдента

1. Вспомните (или посмотрите на Википедии) определение распределения Стьюдента с параметром. Как оно ведёт себя с ростом параметра? Что будет происходить с p-value критериев проверки нормальности и с гипотезой о нормальности с ростом параметра (будет ли она отвергаться чаще или нет)?

__Ответ:__

$T_n \sim \frac{\xi}{\sqrt{\frac{\eta_1 + \dots + \eta_n}{n}}}$,

где $\xi \sim \mathcal{N}(0,1), \eta_1, \dots, \eta_n \sim \chi^2$, независимые,

$\frac{\eta_1 + \dots + \eta_n}{n} \overset{\text{п.в.}}{\rightarrow} 1$ по УЗБЧ, поэтому $\frac{\xi}{\sqrt{\frac{\eta_1 + \dots + \eta_n}{n}}} \overset{\text{п.в.}}{\rightarrow} \xi \sim \mathcal{N}(0, 1)$

С ростом $n$ хвосты Стьюдента уменьшаются. Так например, для $n=1$ распределение Стьюдента совпадает с распределением Коши, у которого даже нет матожидания!

Поэтому с ростом $n$ гипотеза о нормальности будет отвергаться реже, то есть pvalue будет увеличиваться.


##Критерий согласия Колмогорова.

2.1 Вспомните прошлые домашки и сформулируйте теорему Гливенко-Кантелли, а именно часть о поведении статистики $\sqrt{n}D_n$. Постройте по этой теореме критерий согласия Колмогорова вида: если некоторая выборочная статистика попадает в некоторый интервал с конечными (квантили распределения Колмогорова) или бесконечными границами, то гипотеза отвергается.

Теорема: $\sqrt{n}D_n \overset{d}{\rightarrow} K$,

где $K$ -- распределение Колмогорова,
$D_n = \sup\limits_x |F^*_n(x) - F_0(x)|$

Как было показано на семинаре, $D_n = \max\limits_{i = 1, \dots, n} \max \{ \frac{i}{n} - F_0(X_{(i)}), F_0(X_{(i)}) - \frac{i-1}{n}\}$

Критерий выглядит следующим образом:

$S = \{\sqrt{n} D_n(X) > k_{1 - \alpha}\}$, где $k_{1 - \alpha}$ -- квантиль распределения Колмогорова (взяли именно такую, т.к. требуется уровень значимости $\alpha$).

2.2

Промоделируйте поведение статистики $\sqrt{n}D_n$ для некоторого выбранного Вами распределения и достаточно большого (1000+) n. Посчитайте выборочные квантили, которые Вам потребуются для критерия. Опишите теоретически критерий Колмогорова (описал выше).


```python
ALPHA = 0.05


class Kolmogorov:
  def compute_statistics(self, X, hypothetical_cdf):
    X = np.sort(X, axis=1)
    n = X.shape[1]
    F_X = hypothetical_cdf(X)

    max1 = np.max(np.arange(1, n+1) / n - F_X, axis=1)
    max2 = np.max(F_X - np.arange(0,n) / n, axis=1)

    Dn = np.max(np.vstack((max1, max2)), axis=0)

    return np.sqrt(n) * Dn

  def __str__(self):
    return "Колмогоров"
```

```python
criteria1 = [Kolmogorov()]
```

```python
def train_threshold_and_plot_statistics(criteria, num_samples, sample_size):
  plt.figure(figsize=(12, 5))

  for i, criterion in enumerate(criteria, start=1):
    norm_cdf_norm_samples_statistics = criterion.compute_statistics(
      X=sps.norm.rvs(size=(num_samples, sample_size), random_state=1),
      hypothetical_cdf=sps.norm.cdf
    )

    criterion.threshold = np.quantile(norm_cdf_norm_samples_statistics, 1 - ALPHA)

    plt.subplot(1, len(criteria), i)
    plt.title(str(criterion))
    plt.hist(norm_cdf_norm_samples_statistics, bins=40, density=True)
    plt.axvline(x=criterion.threshold, color="red", label=r"$u_{(" + str(1-ALPHA) + r")} \approx$" + f"{criterion.threshold:.3f}")
    plt.legend()

  plt.suptitle("Распределение статистики")
  plt.show();


train_threshold_and_plot_statistics(criteria1, num_samples=14_000, sample_size=14_000)
```

2.3 Сгенерируйте 10000 нормальных выборок размера 100 и посмотрите, как часто критерий отвергает гипотезу.

```python
def research_type_one_errors(criteria, num_samples, sample_size):
  for criterion in criteria:
    norm_cdf_norm_samples_statistics = criterion.compute_statistics(
      X=sps.norm.rvs(size=(num_samples, sample_size), random_state=2),
      hypothetical_cdf=sps.norm.cdf
    )

    reject_ratio = np.mean(norm_cdf_norm_samples_statistics > criterion.threshold)
    print(f"{str(criterion)}, процент ошибок первого рода: {reject_ratio * 100:.2f}%")


research_type_one_errors(criteria1, num_samples=20_000, sample_size=100)
```

__Вывод__

Получили менее $5$% ошибок первого рода, как и хотели!

2.4 Постройте график мощности критерия на альтернативе в виде распределения Стьюдента в зависимости от параметра $k$. То есть для каждого $k$ посчитайте долю отвергаемых критерием стьюдентовских выборок

```python
def research_reject_ratios(criteria, num_samples, sample_size):
  plt.figure(figsize=(10, 5))

  for criterion in criteria:
    reject_ratios = []

    for k in range(1, 30):
      norm_cdf_student_samples_statistics = criterion.compute_statistics(
          X=sps.t.rvs(df=k, size=(num_samples, sample_size), random_state=3),
          hypothetical_cdf=sps.norm.cdf
      )

      reject_ratios.append(
          np.mean(norm_cdf_student_samples_statistics > criterion.threshold))

    plt.plot(reject_ratios, label=str(criterion))

  plt.xlabel("k")
  plt.legend()
  plt.suptitle(r"Доля отвергаемых критерием стьюдентовских выборок $\sim T_k$")
  plt.show();


research_reject_ratios(criteria1, num_samples=20_000, sample_size=100)
```

__Вывод__

Доля отвергаемых критерием стьюдентовских выборок падает, что ожидалось и было описано в пункте 1.

3. Проделайте действия из первой части задачи для [критерия Андерсона-Дарлинга](https://en.wikipedia.org/wiki/Anderson–Darling_test) и [Крамера-фон Мизеса-Смирнова](https://en.wikipedia.org/wiki/Cramér–von_Mises_criterion). Сравните их отвергающую способность с аналогичной у критерия Колмогорова.

```python
class AndersonDarling:
  def compute_statistics(self, X, hypothetical_cdf):
    X = np.sort(X, axis=1)
    n = X.shape[1]
    F_X = hypothetical_cdf(X)

    coef = (2 * np.arange(1,n+1) - 1) / n
    term1 = np.log(F_X)
    term2 = np.log(1 - np.flip(F_X, axis=1))
    S = np.sum(coef * (term1 + term2), axis=1)
    A2 = -n - S
    return A2

  def __str__(self):
    return "Андерсон-Дарлинг"


class CramerVonMises:
  def compute_statistics(self, X, hypothetical_cdf):
    X = np.sort(X, axis=1)
    n = X.shape[1]

    term1 = (2 * np.arange(1,n+1) - 1) / (2 * n)
    term2 = hypothetical_cdf(X)

    nw2 = 1/(12 * n) + np.sum(np.square(term1 - term2), axis=1)
    return nw2

  def __str__(self):
    return "Крамер-фон Мизес"

criteria2 = criteria1 + [AndersonDarling(), CramerVonMises()]
```

### Критерий согласия Пирсона

4.1 Изучите критерий согласия Пирсона. В оригинальной постановке он работает для дискретных распределений. Предложите способ проверки при помощи него гипотез согласия для непрерывных распределений.

__Ответ:__

Можно нарезать носитель плотности на бины. И далее работать как с дискретным (попало ли в i-й бин).

Как сказали на семинаре, важное условие, что для каждого бина матожидание кол-ва попаданий в него (за $n$ экспериментов) должно быть хотя бы 5 ($n p^0_i \geq 5$) и должно быть хотя бы 50 экспериментов ($n \geq 50$).

Если распределение известное (например, нормальное) и у него известны квантили (хотя бы примерно), можно разбить носитель на $m$ отрезков, в каждый из которых нормальное распр-е попадает с примерно одинаковой вероятностью (то есть в среднем в каждый должно попасть $\approx \frac{n}{m}$ экспериментов) и взять их в качестве бинов.

4.2 Постройте критерий согласия Пирсона для нормальных распределений. Опишите, какие квантили для распределения хи-квадрат Вам потребуются.

Ответ:

Делаем как описано выше, то есть

$m$ -- кол-во бинов

$\mu_i = \sum \limits_{j=1}^n I_{\big(u_{\frac{i-1}{m}}, u_{\frac{i}{m}}\big]}(X_j)$ -- сколько экспериментов попало в i-й бин

$\widehat{\chi} = \sum \limits_{i=1}^m \frac{(\mu_i - \frac{n}{m})^2}{\frac{n}{m}}$

$H_0: \widehat{\chi} \overset{d}{\rightarrow} \chi^2_{m-1}$

$H_1: \widehat{\chi} \overset{d}{\rightarrow} \infty$

Поэтому критерий, $S = \{\widehat{\chi} > u^{\chi^2_{m-1}}_{1-\alpha}\}$, то есть нужна $1-\alpha$ квантиль хи-квадрат с $m-1$ степенями свободы.

```python
class PearsonForNormality:
  def __init__(self, num_bins):
    self.num_bins = num_bins

  def compute_statistics(self, X, hypothetical_cdf=sps.norm.cdf):
    if hypothetical_cdf != sps.norm.cdf:
      raise Exception("Normal distribution cdf expected")

    n = X.shape[1]

    X = X.reshape(-1, n, 1)

    borders = sps.norm.ppf(np.arange(self.num_bins+1) / self.num_bins)
    mu = np.sum((X <= borders[1:]) & (borders[:-1] < X), axis=1)

    expected_hitcount = n / self.num_bins

    chi_hat = np.sum(np.square(mu - expected_hitcount), axis=1) / expected_hitcount

    return chi_hat

  def __str__(self):
    return f"Пирсон, кол-во бинов: {self.num_bins}"


pearson_criteria = [PearsonForNormality(num_bins) for num_bins in [10, 20, 30]]
criteria3 = criteria2 + pearson_criteria
```

4.3. Промоделируйте статистику хи-квадрат, считая доступным только нормальное распределение (другие семплировать нельзя!).

```python
for criterion in tqdm(pearson_criteria):
  norm_sample = sps.norm.rvs(size=(20_000, criterion.num_bins-1), random_state=42)
  chi_square_sample = np.sum(np.square(norm_sample), axis=1)
  criterion.threshold = np.quantile(chi_square_sample, 1-ALPHA)

plt.title(r"Распределение $\chi^2_{" + str(pearson_criteria[-1].num_bins) + r"}$")
plt.hist(chi_square_sample, bins=30)
plt.axvline(x=pearson_criteria[-1].threshold, color="red", label=r"$u_{(" + str(1-ALPHA) + r")} \approx$" + f"{pearson_criteria[-1].threshold:.3f}")
plt.legend()
plt.show();
```

### Стандартные способы проверки

5.1 Изучите критерий Шапиро-Уилка по [документации](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html#scipy.stats.shapiro).

Что он проверяет? Будет ли он корректен для проверки распределения на равенство $\mathcal N(0, 1)$?

Ответ:

Критерий используется для проверки нулевой гипотезы $H_0$: "случайная величина X распределена нормально", поэтому втупую его нельзя использовать для проверки на __равенство__, то есть нулевой гипотезы "$X \sim \mathcal{N}(0, 1)$", но можно комбинировать его с мощными критериями, которые работают в предположении, что $X \sim N(a, \sigma^2)$

Проделайте пункты, аналогичные 2.3 и 2.4 для этого критерия.

```python
class Shapiro:
  def compute_statistics(self, X, hypothetical_cdf=sps.norm.cdf):
    if hypothetical_cdf != sps.norm.cdf:
      raise Exception("Normal distribution cdf expected")

    # Возвращаем минус, чтобы получился критерий вида T(X) > u
    return -sps.shapiro(X, axis=1).statistic

  def __str__(self):
    return "Шапиро"

shapiro = Shapiro()
criteria5 = criteria4 + [shapiro]
```

Какой-то очевидной зависимости между разными колонками не наблюдается.

Посчитайте коэффициенты корреляции Пирсона и Спирмена. Отличаются ли они? Почему?

```python
class Pearson:
  def __init__(self, alpha):
    self.alpha = alpha

  def coeff(self, X, Y):
    mu_X = np.mean(X)
    mu_Y = np.mean(Y)

    return (np.mean(X * Y) - mu_X * mu_Y) / (np.mean(X**2) - mu_X**2)**0.5 / (np.mean(Y**2) - mu_Y**2)**0.5

  def statistic(self, X, Y):
    r = self.coeff(X, Y)
    n = len(X)

    return r * (n-2)**0.5 / (1-r**2)**0.5

  def rejects(self, X, Y):
    n = len(X)
    threshold = sps.t.ppf(1 - self.alpha, df=n-2)
    return self.statistic(X, Y) > threshold


class Spearman:
  def __init__(self, alpha):
    self.alpha = alpha

  def coeff(self, X, Y):
    # Ранги нумеруются с единицы!
    X_ranks = np.argsort(X)+1
    Y_ranks = np.argsort(Y)+1

    return 1 - 6 / (n**3 - n) * np.sum((X_ranks - Y_ranks)**2)

  def statistic(self, X, Y):
    rho = self.coeff(X, Y)
    n = len(X)

    return abs(0.5 * rho * ((n-1)**0.5 + (n-2)**0.5/(1-rho**2)**0.5))

  def rejects(self, X, Y):
    n = len(X)
    threshold = 0.5 * (sps.norm.ppf(1 - 0.5 * self.alpha) + sps.t.ppf(1 - 0.5 * self.alpha, df=n-2))

    return self.statistic(X, Y) > threshold


criteria = [Pearson(0.1), Spearman(0.1)]
```

Но отметим, что коэффициенты получились одинакового знака, это здорово!

__Решение__

Поставим нулевую гипотезу для каждой пары выборок:

То есть 3 гипотезы, посмотрим, что с ними произойдет (будут ли они отвергнуты)

Будем использовать критерии основанные на коэффициентах Пирсона и Спирмена, которые рассказали на семинаре по матстатам, а именно.

$S_{\text{Пирсон}} = \Big\{r_{X,Y} \frac{\sqrt{n-2}}{\sqrt{1-r^2_{X,Y}}} > t^{n-2}_{1-\alpha} \Big\}$, где $r_{X,Y}$ -- сам коэффициент

$S_{\text{Спирман}} = \Big \{\bigr|\frac{1}{2}\rho_s \big( \sqrt{n-1} + \frac{\sqrt{n-2}}{\sqrt{1 - \rho^2_s}}\big)\bigr| > \frac{1}{2}\big(z_{1 - \frac{\alpha}{2}} + t^{n-2}_{1 - \frac{\alpha}{2}}\big)\Big \}$, где $\rho_s$ -- коэффициент Спирмана

Будем строить критерии уровня значимости $0.1$, потому что выборка небольшая и эффект трудно назвать сильным, при $\alpha=0.05$ он может не отловиться (то есть $\alpha = 0.1$).

__Ответ:__

Я считаю, что в такой постановке нужно принимать решение консервативно: если один из критериев отверг гипотезу, значит ее в целом строит отвергнуть, потому что разные критерии это способы взглянуть на выборку под разными углами, и если "под некоторым углом" нашли несостыковку, значит в целом есть несостыковка (типа redflag).

Единственный момент, если у нас $m$ критериев и у каждого вероятность допустить ошибку первого рода не превосходит $\alpha$, тогда вероятность что никто из кретериев не допустит ошибку первого рода хотя бы $(1 - \alpha)^m$, то есть у композитного критерия вероятность ошибки первого рода не превосходит $1 - (1 - \alpha)^m$, поэтому если мы хотим композитный критерий уровня значимости $\alpha$, нужно корректировать уровни значимости для примарных критериев.

На практике это называется **множественная проверка гипотез**. Можете поискать об этом информацию и написать поправку Бонферонни или любую другую. Эта часть не оценивается, но может быть интересна:

```python
class Composite:
  def __init__(self, alpha, primary_criteria_cls):
    m = len(primary_criteria_cls)
    primary_alpha = 1 - (1 - alpha)**(1/m)
    self.primary_criteria = [criterion_cls(primary_alpha) for criterion_cls in primary_criteria_cls]
    self.alpha = alpha

  def rejects(self, X, Y):
    return any([criterion.rejects(X, Y) for criterion in self.primary_criteria])

composite_criterion = Composite(0.1, [Pearson, Spearman])
```

## Задача 4. T-критерий Стьюдента

*Первые три пункта стоят по 5 баллов, последний, бонусный, --- 10 баллов.*

Распространённой задачей статистики является обнаружение эффекта: есть две (или более, но мы рассмотрим случай с двумя) выборки, правда ли, что их средние отличаются? Классическим критерием для проверки такой гипотезы является *t-критерий*. Напомним, как он выводится.

Пусть $X_1, \ldots, X_n \sim \mathcal{N}(\mu_1, \sigma^2)$ и $Y_1, \ldots, Y_m \sim \mathcal{N}(\mu_2, \sigma^2)$ --- две независимые выборки (обратите внимание, мы предполагаем, что их дисперсии равны, хоть мы её и не знаем). На семинаре мы убедились, что при верности нулевой гипотезы $H_0\colon \mu_1 = \mu_2$ справедливо
$$
T(\mathbf X, \mathbf Y) = \sqrt{\frac{nm}{n+m}} \cdot \frac{\overline{X} - \overline{Y}}{\sqrt{\frac{(n-1)s^2(\mathbf X) + (m-1)s^2(\mathbf Y)}{n+m-2}}} \sim T_{n+m-2},
$$
где $T_k$ --- распределение Стьюдента с $k$ степенями свободы, $s^2(\mathbf X)$ и $s^2(\mathbf Y)$ --- *несмещённые* выборочные дисперсии. Отсюда мы получаем критерий Стьюдента вида $R_{\alpha} = \{(\mathbf x, \mathbf y)\colon |T(\mathbf x, \mathbf y)| > t_{n+m-2, 1-\alpha/2}\}$ уровня значимости $\alpha$ (здесь $t_{k,p}$ --- $p$-квантиль распределения $T_k$).

1. В общем случае, когда про распределение выборок мы ничего не знаем, а гипотезу о равенстве средних проверить хочется, применяют асимптотический *критерий Вальда*, основанный на асимпт. нормальной оценке разности средних:
$$
W(\mathbf X, \mathbf Y) = \frac{\overline{X} - \overline{Y}}{\sqrt{s^2(\mathbf X)/n + s^2(\mathbf Y)/m}} \stackrel{d}{\to} \mathcal{N}(0, 1).
$$

Универсальность -- вещь хорошая, но тут распределение статистики известно лишь в пределе, что может вредить ошибке I рода. Убедимся в этом, ~провальдедируя~ провалидируя критерии.

Насэмплируйте достаточно много (в идеале 100000, но можно и чуть меньше, если мощности не позволяют) пар нормальных выборок размера 200 из нулевой гипотезы.

```python
X, Y = sps.norm.rvs(loc=1, scale=2, size=(2, 100_000, 200), random_state=5)
```

Посчитайте статистики критериев Вальда и Стьюдента для всех префиксов вместе с соответствующими pvalue. Для второго критерия можете воспользоваться методом [`scipy.stats.ttest_ind`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html).

__Анализ для кртерия Вальда__

$\mathbb{P}_0$ -- $W(X, Y) \sim \mathcal{N}(0, 1)$

$\Longrightarrow$ p-value$(\widetilde{X}, \widetilde{Y})$ = $\mathbb{P}_0 \Big(W(X, Y) > W(\widetilde{X}, \widetilde{Y}) \Big) = 1 - \Phi\big( W(\widetilde{X}, \widetilde{Y})\big)$, где $\Phi$ -- ф.р. стандартного нормального

```python
def cummean(X):
  return np.cumsum(X, axis=-1) / np.arange(1, X.shape[-1]+1)


def s2(X):
  n = np.arange(1, X.shape[-1]+1)
  return n/(n-1) * (cummean(X**2) - cummean(X)**2)


def wald_test_all_prefixes(X, Y):
  n = np.arange(1, X.shape[-1]+1)
  m = np.arange(1, Y.shape[-1]+1)

  W = (cummean(X) - cummean(Y)) / (s2(X)/m + s2(Y)/n)**0.5

  pvalues = 1 - sps.norm.cdf(W)

  return W, pvalues


def t_test_all_prefixes(X, Y):
  n = np.arange(1, X.shape[-1]+1)
  m = np.arange(1, Y.shape[-1]+1)

  T = (n*m / (n+m))**0.5 * (cummean(X) - cummean(Y)) / ((n-1)*s2(X)/(n+m-2) + (m-1)*s2(Y)/(n+m-2))**0.5

  pvalues = 1 - sps.t.cdf(T, df=n+m-2)

  return T, pvalues


wald_statistics, wald_pvalues = wald_test_all_prefixes(X, Y)
t_statistics, t_pvalues = t_test_all_prefixes(X, Y)
```

Для нескольких (3-4 штуки хватит) префиксов визуализируйте распределения pvalue. Что можно по ним сказать?

```python
for pvalues, title in zip([wald_pvalues, t_pvalues],
                          ["Вальд", "Стьюдент"]):

  sns.displot({f"n={prefix_size}": pvalues[:, prefix_size-1]
               for prefix_size in [5, 50, 100]}, kind="kde")

  plt.title(f"{title}, распределение p-value")
```

4. *Это бонусный пункт*.

Как мы поняли ранее, для распределений, отличных от нормального, распределение pvalue при равенстве средних может быть далёким от равномерного. Можно, конечно, сдвинуть пороговое значение при проверки критерия, чтобы подправить вероятность ошибки I рода, но будет ли такой критерий достаточно мощным? Не стоит ли рассмотреть какую-нибудь другой критерий? В этом пункте мы рассмотрим такой, он называется *критерием отношения правдоподобий* (или сокращённо КОП).

Рассмотрим выборку $X_1, \ldots, X_n$, которая пришла из распределения $\mathsf{P}_{\theta}$ с совместной плотностью $\rho_{\theta}(\mathbf x)$, параметризованного неким $\theta \in \Theta \subset{R^k}$. Поставим на проверку гипотезу $H_0\colon \theta \in \Theta_0$, где $\Theta_0 \subset \Theta$ --- поверхность размерности $l$ (простой пример такой гипотезы -- $H_0\colon \theta_1 = \ldots = \theta_{k-l} = 0$). КОП предлагает рассмотреть статистику
$$
LR(\mathbf X) = \frac{\sup_{\theta \in \Theta} \rho_{\theta}(\mathbf X)}{\sup_{\theta \in \Theta_0} \rho_{\theta}(\mathbf X)}.
$$

Своего рода она показывает, насколько общее правдоподобие выше частного, и если разрыв между ними очень большой, то критерий будет отвергать гипотезу $H_0$. Оказывается, что в некоторых условиях регулярности справедлив следующий асимптотический результат, называемый теоремой Уилкса:
$$
2 \ln LR(\mathbf X) \stackrel{d}{\to} \chi^2_{k-l}.
$$

Таким образом, критерий вида $\{\mathbf x\colon 2 \ln LR(\mathbf x) > \chi^2_{k-l,1-\alpha}\}$ будет иметь асимптотический уровень значимости $\alpha$. Применим его к проверке равенства средних, но для совсем другого семейства распределений, например, Коши.
